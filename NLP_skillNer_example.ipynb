```python
# skill extraction with skillNer

import en_core_web_lg
from spacy.matcher import PhraseMatcher

# import skill extractor
from skillNer.skill_extractor_class import SkillExtractor
from skillNer.general_params import SKILL_DB

# init params of skill extractor
nlp = en_core_web_lg.load()
# init skill extractor
skill_extractor = SkillExtractor(nlp, SKILL_DB, PhraseMatcher)
```

```python
# extract skills from job_description
job_description = """
You are a Python developer with a solid experience in web development
and can manage projects. You quickly adapt to new environments 
and speak fluently English and French
"""

annotations = skill_extractor.annotate(job_description)
```

```python
# inspect annotations
skill_extractor.describe(annotations)
```

```python
display(annotations)
```

```python
# Extracting values for 'doc_node_value' key
doc_node_values = [item['doc_node_value'] for item in annotations['results']['full_matches']]   # item returns a list of tuples with the key, value pairs
doc_node_values.extend([item['doc_node_value'] for item in annotations['results']['ngram_scored']]) # extend concatenates a list with another list

print(doc_node_values)
```
